```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer, TextDataset, DataCollatorForLanguageModeling, Trainer, TrainingArguments
```

```python
def fine_tune_gpt2(dataset_path, model_name='gpt2', output_dir='./fine_tuned_gpt2', epochs=5, batch_size=16):
    # Load the pre-trained model and tokenizer
    model = GPT2LMHeadModel.from_pretrained(model_name)
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)

    # Prepare the dataset
    dataset = TextDataset(
        tokenizer=tokenizer,
        file_path=dataset_path,
        block_size=128  # Adjust based on the length of your stories
    )

    # Prepare the data collator
    data_collator = DataCollatorForLanguageModeling(
        tokenizer=tokenizer,
        mlm=False  # Masked language modeling is not applicable for GPT-2
    )

    # Set up training arguments
    training_args = TrainingArguments(
        output_dir=output_dir,
        overwrite_output_dir=True,
        num_train_epochs=epochs,
        per_device_train_batch_size=batch_size,
        save_steps=10_000,
        save_total_limit=2,
        logging_steps=1000,
    )

    # Initialize the Trainer
    trainer = Trainer(
        model=model,
        args=training_args,
        data_collator=data_collator,
        train_dataset=dataset,
    )

    # Fine-tune the model
    trainer.train()

    # Save the model
    trainer.save_model(output_dir)
    tokenizer.save_pretrained(output_dir)

# Fine-tune the model using your dataset
dataset_path = '/content/fairy-tales/merged_clean.txt'
fine_tune_gpt2(dataset_path)
```

```python
def load_fine_tuned_model(output_dir='./fine_tuned_gpt2'):
    model = GPT2LMHeadModel.from_pretrained(output_dir)
    tokenizer = GPT2Tokenizer.from_pretrained(output_dir)
    return model, tokenizer
```

```python
def generate_story(prompt, model, tokenizer, max_length=500, temperature=0.7):
    # Encode the prompt text
    input_ids = tokenizer.encode(prompt, return_tensors='pt')

    # Generate text
    with torch.no_grad():
        output = model.generate(
            input_ids,
            max_length=max_length,
            temperature=temperature,
            num_return_sequences=1,
            no_repeat_ngram_size=2,
            top_k=50,
            top_p=0.95,
            do_sample=True,
            eos_token_id=tokenizer.eos_token_id
        )

    # Decode the generated text
    story = tokenizer.decode(output[0], skip_special_tokens=True)
    return story

# Load the fine-tuned model
model, tokenizer = load_fine_tuned_model()
```

```python
# Generate a story with the fine-tuned model
prompt = "Once upon a time"
story = generate_story(prompt, model, tokenizer)
print(story)
```
